## Method 1: careerjet API

Documentation: https://github.com/careerjet/careerjet-api-client-python
Register as publisher, get affid. https://www.careerjet.com/partners/

Drawbacks: It seems that in the returned DataFrame, the job description is truncated so you won't be able to filter the keywords of requirements or skillset, etc. Please think about how to obtain the entire job description.  

```python
from careerjet_api_client import CareerjetAPIClient

cj  =  CareerjetAPIClient("en_US");

result = cj.search({
                        'location'    : 'new york city',
                        'contracttype': 'p',
                        'keywords'    : 'data analyst',
                        'affid'       : '7a4b04c81d8efbc78be88a8135d28913',
                        'user_ip'     : '134.154.75.228',
                        'url'         : 'http://www.example.com/jobsearch?q=python&l=london',
                        'user_agent'  : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:63.0) Gecko/20100101 Firefox/63.0'
                      });
result
```

## Method 2: Scraping directly from github

You may change the last part in url ``"description=data+analyst&location="`` and try different combination. 

Drawbacks: The job posting in github might not be up-to-date and data amount might be very limited. (But the desciption is full). 

```python
import urllib
from bs4 import BeautifulSoup as BS
import json


url = "https://jobs.github.com/positions.json?description=data+analyst&location="

text = urllib.urlopen(url)

data = json.load(text)

print data[0]['description']

```
